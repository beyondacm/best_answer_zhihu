```python
def readlist(file_name) :
    with open(file_name, 'r') as f :
        line = f.readline()
        # print line
        len_list = line.split(' ') 
        len_list = map(int, len_list)
        # print len_list
    
    return len_list
```

```python
# IMPORTANT THIS MAYBE CHANGE
def Split_Final_Test_Set(Total_Final, test_len, final_test) :
    
    start = end = 0
    for i in range( len(test_len) ) :
        end += test_len[i]
        print (i, start, end)
        test_i = Total_Final[start:end]
        # test_i = (test_i - similar01_best.min()) / (similar01_best.max() - similar01_best.min())
        # test_i.to_csv('../TOTAL_DATA_FEATURES/TEST/' + 'test_' + str(i) + '.csv', encoding='utf-8', index = False)
        test_i.to_csv('../MODEL/SVM_Data/' + LEVEL + final_test + '/test_' + str(i) + '.csv', encoding='utf-8', index = False)
        test_i.drop(test_i.columns[[1, 2, 3]], inplace=True, axis=1)
        test_i.to_csv('../MODEL/MODEL_Data/' + LEVEL + final_test + '/test_' + str(i) + '.csv', encoding='utf-8', index=False, header=False)
        start += test_len[i]
```

```python
# 1000x5 : 1000 question, each question find 5 similar question
# 1000x5/Final_Test_00 : Divide 1000 question to 10 folds , each fold has 100 question (about 600-700 Q-A pair)
# 1000x5/Final_Test_00 : has 100 question, seperate each question 
#                        to ../MODEL/SVM_DATA/Final_Test_00/test_0 - test_99.csv for testing 
for i in range(10) :
    final_test = 'Final_Test_0' + str(i)
    final_test_file = 'Final_Test_0' + str(i) + '.csv'
    len_file = 'Len_0' + str(i) + '.txt'
    
    # print PATH_FINAL_TEST_SET + test_features
    # print PATH_FINAL_TEST_SET + test_len
    Total_Final = pd.read_csv( PATH_FINAL_TEST_FEATURES + final_test_file )
    test_len = readlist( PATH_FINAL_TEST_SET + len_file )
    print sum(test_len)
    Split_Final_Test_Set(Total_Final, test_len, final_test)
```
